# Style
[Customizing Matplotlib: Configurations and Stylesheets](https://jakevdp.github.io/PythonDataScienceHandbook/04.11-settings-and-stylesheets.html)

````
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
````

````
#plt.style.use('seaborn-dark') # this will change the style for the rest of the session
# OR
with plt.style.context('seaborn-dark'):
    hist_and_lines()
````

````
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
````

````
with plt.style.context('ggplot'):
    hist_and_lines()
````

# subplots() VS. subplot()


*   subplots `MATLAB-style Interface`
*   subplot  `Object-oriented interface`



## 1. subplot()


*   this interface is stateful
*   keeps track of the "current" figure and axes
*   all plt commands are applied
*   get a reference to these using the `plt.gcf()` (get current figure) and `plt.gca()` (get current axes) routines.


Pros:
* this stateful interface is fast and convenient for simple plots

Cons:
* always need to remember an "active" figure or axes
* once the second panel is created, how can we go back and add something to the first?

````
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

## 2. subplots() `Object-oriented interface`
>  available for these more complicated situations
>  and for when you want more control over your figure
<br/>

the plotting functions are methods of explicit Figure and Axes objects. To re-create the previous plot using this style of plotting, you might do the following:

> plt.subplots() is a function that returns a tuple containing a figure and axes object(s)
<br>

````
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

````
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows=2, ncols=2)

fig, [ax1, ax2, ax3, ax4] = plt.subplots(nrows=1, ncols=4)
````

In most cases, the difference is as small as switching plt.plot() to ax.plot().

# Axes VS. Subplot

可以把fig想象成windows的桌面，你可以有好几个桌面。然后axes就是桌面上的图标，subplot也是图标
<br/>
他们的区别在：axes是自由摆放的图标，甚至可以相互重叠，而subplot是“自动对齐到网格”。但他们本质上都是图标
<br>
也就是说subplot内部其实也是调用的axes，只不过规范了各个axes的排列罢了。

## 1. subplot:
> is just a special case of an Axes that lives on a regular rows by columns grid of Subplot instances
<br>
前两个数字代表要生成几行几列的子图矩阵
数字代表选中的子图位置

````
fig = plt.figure()  #Create 画板
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
plt.show()
````

## 2.axes`灵活子图`:
用add_axes()方法生成一个轴域（Axes）
<br/>
括号里面的值前两个是轴域原点坐标（从左下角计算的）
<br/>
后两个是显示坐标轴的长度

````
fig = plt.figure()
ax3 = fig.add_axes([0.1,0.1,0.8,0.8])
ax4 = fig.add_axes([0.72,0.72,0.16,0.16])
plt.show()
````

# Axes VS. pyplot

*  ax.plot() `函数式编程`
* plt.plot() `对象式编程` 

````
# plt (pyplot)  函数式编程
# Pros:
# 表达简洁

# Cons:
# 没有体现出真正画图的实现过程(没有搞清楚Figure Axes Subplot 等概念的时候，依然可以轻松的用pyplot函数画图)
# 当子图较多的时候，对子图的操作容易陷入混乱，因为从代码上并不能字节观察出到底在操作那张子图
plt.plot([1,2,3,4],[10,20,25,30],color = 'lightblue', linewidth=3)
plt.xlim(0.5,4.5)
````

````
# ax(axes)
# Pros:
# 表达明确，分步生成 Figure 和 Axes/Subplot
# 作过程直接可以看出是在那张子图上操作。

# Cons:
# 缺点就是，需要写的代码比较多，不够简洁。
fig=plt.figure()
ax = fig.add_subplot(111)

print(type(ax))

ax.plot([1,2,3,4],[10,20,25,30],color = 'lightblue', linewidth=3)
ax.set_xlim(0.5,4.5)
````

> “明了胜于晦涩”（Explicit is better than implicit）
> “简洁胜于复杂”（Complex is better than complicated）

所以，还是看你的使用场景，假如不需要画子图的时候，用一用简单的pyplot方法也没什么不好。
<br/>
但是初学者最好还是能够坚持先使用Axes对象属性的方法，这样对于画图的实现过程可以加深理解。

另外：

`fig, ax = plt.subplots()` 

is more concise than 

```
fig = plt.figure()
ax = fig.add_subplot(111)
```



#就近原则

就近原则：在每个函数操作时对上一层资源的依赖遵守最近原则，举例如下：



* subplot()作用在最近使用的画图板上，有可能是缺省的画图板figure(1)。
* axes()作用在最近使用的画图板上，有可能是缺省的画图板figure(1)。
* axis()作用在最近使用的图形区上，可能来自subplot、axes等
* plot()作用在最近使用的图形区上，可能来自subplot、axes等
